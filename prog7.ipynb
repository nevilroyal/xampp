```python
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_csv("D:/Groceries_dataset.csv")
```

```python
transactions = df.groupby(['Member_number','Date'])
transactions.count()
```

```python
transactions = [i[1]['itemDescription'].tolist() for i in list(transactions)]
transactions[:20]
```

```python
transactions
```

```python
rules = apriori(transactions, min_support=0.001, min_confidence=0.05,
min_lift=1.2, min_length=2, max_length=2)
#rules = apriori(transactions, min_support=0.001, min_length=2, max_length=2)
```

```python
results = list(rules)
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    support = [result[1]*100 for result in results]
    return list(zip(lhs,rhs,support))
    
final_result = pd.DataFrame(inspect(results), columns=['Product 1','Product2','Support'])
final_result['Rule'] = final_result['Product 1'] + '->' + final_result['Product2']
```
