```python
# Plot a bar graph using countplot for Height Distribution of US Presidents
# (Number of Presidents based on the height in ascending order)
plt.figure(figsize=(15, 5))
#heights = data['height(cm)'].value_counts().index
heights = data['height(cm)'].value_counts(ascending=True).index
ax = sns.countplot(data=data, x='height(cm)', order=heights)
abs_values = data['height(cm)'].value_counts(ascending=True).values
ax.bar_label(container=ax.containers[0], labels=abs_values)
plt.xticks(rotation=90)
# Define x-axis range
plt.yticks(np.arange(0, 11, step=1))
plt.title("Height Distribution of US Presidents")
plt.xlabel("Height [cm]")
plt.ylabel("Number of Presidents")
```

```python
# Plot a bar graph using countplot for Height Distribution of US Presidents
# (Number of Presidents based on the height in descending order)
plt.figure(figsize=(15, 5))
ax = sns.countplot(x=data['height(cm)'],order=data['height(cm)'].value_counts(ascending=False).index);
abs_values = data['height(cm)'].value_counts(ascending=False).values
ax.bar_label(container=ax.containers[0], labels=abs_values)
# Define x-axis range
plt.yticks(np.arange(0, 11, step=1))
plt.title("Height Distribution of US Presidents")
plt.xlabel("Height [cm]")
plt.ylabel("Number of Presidents")
plt.show()
```

```python
# Plot a bar graph using countplot for Height Distribution of US Presidents
# along with % of distributions (Number of presidents based on the height in descending order)
plt.figure(figsize=(15, 5))
ax = sns.countplot(x=data['height(cm)'],order=data['height(cm)'].value_counts(ascending=False).index)
abs_values = data['height(cm)'].value_counts(ascending=False).values
rel_values = data['height(cm)'].value_counts(ascending=False, normalize=True).values * 100
#lbls = [f'{p[0]} ({p[1]:.0f}%)' for p in zip(abs_values, rel_values)]
lbls = [f'{p[0]} ({p[1]:.3f}%)' for p in zip(abs_values, rel_values)]
ax.bar_label(container=ax.containers[0], labels=lbls, rotation=90)
#ax.bar_label(container=ax.containers[0], labels=lbls)
# Define x-axis range
plt.yticks(np.arange(0, 12, step=1))
plt.title("Height Distribution of US Presidents (with % of distributions)")
plt.xlabel("Height [cm]")
plt.ylabel("Number of Presidents")
plt.show()
```

```python
#Group and sort the heights of the presidents
sorted_data = data.groupby('height(cm)').count().sort_values(by=['height(cm)'],ascending=False)
#print(sorted_data)
s1=pd.Series(sorted_data.index.values.tolist(),name='height(cm)')
s2=pd.Series(sorted_data['order'].values.tolist(),name='count')
df = pd.concat([s1,s2], axis=1)
print(df)
# create bar graph
ax = df.plot.bar(x = 'height(cm)', y = 'count', color ='darkblue', figsize =(5, 5), legend=True, fontsize=12)
ax.bar_label(ax.containers[0])
plt.xlabel("Height [cm]")
plt.ylabel("Number of Presidents")
plt.title("Height Distribution of US Presidents")
plt.show()

s1=pd.Series(sorted_data.index.values.tolist(),name='height(cm)')
s2=pd.Series(sorted_data['order'].values.tolist(),name='count')
df = pd.concat([s1,s2], axis=1)
ax = df[['height(cm)','count']].plot(kind='bar', x = 'height(cm)', y = 'count',title="Height Distribution of US Presidents", color ='maroon', figsize=(15,10), legend=True, fontsize=12)
ax.set_xlabel("Height [cm]", fontsize=12)
ax.set_ylabel("Number of Presidents", fontsize=12)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(16, 10))
hieghts_dict = {k: int(v) for k, v in zip(df['height(cm)'], df['count'])}
df_new = pd.DataFrame(list(hieghts_dict.items()), columns=['height(cm)','count'])
patches, labels, pct_texts = plt.pie(df_new['count'],labels=df_new['height(cm)'], radius=3, startangle=160,
autopct=lambda p: f"{p:.2f}% ({p *sum(df['count']) / 100:,.0f})",
rotatelabels=True, pctdistance=0.5)
for label, pct_text in zip(labels, pct_texts):
    pct_text.set_rotation(label.get_rotation())
    
labels[0].set_fontsize(12)
plt.legend(loc="upper left")
plt.title('Height Distribution of US Presidents', pad=32)
plt.axis('equal')
plt.tight_layout()
```
